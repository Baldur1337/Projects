```python
# Import Libaries

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
import numpy as np
```

```python
# Read in the csv (data)
df = pd.read_csv(r"D:\Benutzer\Christian\Python\Project\kaggle_movies\movies.csv")
```

```python
# Get to known the data
df.head()
```

```python
# Identify missing Data
for column in df.columns:
    pct_missing = df[column].isnull().mean()
    print(f"In column '{column}': {round(pct_missing*100,2)} % of data is missing.")
```

```python
# Drop every row with missing data
df = df.dropna()
```

```python
# Drop any dublicates
df.drop_duplicates()
```

```python
# Data types for the columns
df.dtypes
```

```python
# Deletion of column "released" as the data contained therein are often incorrect 
df.drop("released",axis=1)
```

```python
# Sorting data by gross revenue, not changing the original order
df.sort_values(by=["gross"], inplace=False, ascending=False)
```

```python
#Scatter plot budget ~ gross

figure(figsize=(6,4), dpi=150)
plt.scatter(x=df.budget,y=df.gross)

plt.title("Budget ~ Gross Earnings")
plt.xlabel("Gross Earnings")
plt.ylabel("Budget")

plt.show()
```

```python
# Plot "budget vs. gross" using seaborn
figure(figsize=(6,4), dpi=150)
sns.regplot(x="budget",y="gross",data=df,scatter_kws={"color": "white"}, line_kws={"color": "red"})
```

```python
# Start looking at correlation
df.corr(method="pearson")
```

```python
correlation_matrix = df.corr(method="pearson")
figure(figsize=(6,4), dpi=150)
sns.heatmap(correlation_matrix,annot=True)
```

```python
# Changing non-numerical entrys to numerical ones
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == "object"):
        df_numerized[col_name] = df_numerized[col_name].astype("category")
        df_numerized[col_name] = df_numerized[col_name].cat.codes
```

```python
correlation_matrix = df_numerized.corr(method="pearson")
figure(figsize=(6,4), dpi=1500)
sns.heatmap(correlation_matrix,annot=False)
```

```python
correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
# Sorting each correlation
sorted_pairs = corr_pairs.sort_values(ascending=False)

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs > 0.5)]
high_corr = high_corr[(high_corr != 1)] #removes correlation with items itselfs eg. name/name, genre/genre etc.
high_corr.drop_duplicates(inplace=True) #eliminates the double appearance of results eg. budget/gross and gross/budget
```
